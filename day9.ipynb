```python
#10.Write a program to check Weekend or Weekday

```

```python

day = input("Enter a day: ").lower()

weekend = ["saturday", "sunday"]

if day in weekend:
    print("It is Weekend")
elif day in ["monday", "tuesday", "wednesday", "thursday", "friday"]:
    print("It is Weekday")
else:
    print("Invalid day entered")
```

```python
mark = int(input("Enter your mark: "))

if mark >= 90 and mark <= 100:
    print("Grade: A")
elif mark >= 80:
    print("Grade: B")
elif mark >= 70:
    print("Grade: C")
elif mark >= 60:
    print("Grade: D")
elif mark >= 0:
    print("Grade: Fail")
else:
    print("Invalid mark")

  Enter your mark: 95
  Grade: A  
```

```python
#D2. Vowel or Consonant with sample input/output.
```

```python
letter = input("Enter a letter: ")

if len(letter) == 1 and letter.isalpha():
    if letter.lower() in ['a', 'e', 'i', 'o', 'u']:
        print("It is a Vowel")
    else:
        print("It is a Consonant")
else:
    print("Invalid input")


Enter a letter: a
It is a Vowel
```
